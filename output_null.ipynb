```python
%load_ext autoreload
%autoreload 2

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df1 = pd.read_csv('ozel/scores.dat', sep=' ', header=3, names=['mag', 'aspin', 'inc', 'Rhigh', 'score'])
df2 = pd.read_csv('cache/Illinois_thermal_w3/null.tsv', sep='\t')
df3 = pd.read_csv('cache/Illinois_thermal_w4/null.tsv', sep='\t')
df4 = pd.read_csv('cache/Illinois_thermal_w5/null.tsv', sep='\t')
```

```python
df12 = pd.merge(df1, df2, on=['mag', 'aspin', 'Rhigh', 'inc'])
df13 = pd.merge(df1, df3, on=['mag', 'aspin', 'Rhigh', 'inc'])
df14 = pd.merge(df1, df4, on=['mag', 'aspin', 'Rhigh', 'inc'])
```

```python
Y = np.maximum(df12.score_y, df12.score_y)

plt.scatter(df12.score_x, Y)
plt.gca().set_aspect('equal')
plt.xlabel('ozel')
plt.ylabel('new')
```

```python
Y = np.maximum(df13.score_y, df13.score_y)

plt.scatter(df13.score_x, Y)
plt.gca().set_aspect('equal')
plt.xlabel('ozel')
plt.ylabel('new')
```

```python
Y = np.maximum(df14.score_y, df14.score_y)

plt.scatter(df14.score_x, Y)
plt.gca().set_aspect('equal')
plt.xlabel('ozel')
plt.ylabel('new')
```

```python
Y = np.maximum(np.maximum(df12.score_y, df13.score_y), df14.score_y)

plt.scatter(df13.score_x, Y)
plt.gca().set_aspect('equal')
plt.xlabel('ozel')
plt.ylabel('new')
```
